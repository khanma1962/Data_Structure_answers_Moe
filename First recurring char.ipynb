```
def func(mylist):

  for i in range(0,len(mylist)):
    for j in range(i+1,len(mylist)):
      if mylist[i] == mylist[j]:
        return mylist[i] 
  return 0

def hashtable(mylist):
  mydict = {}
  for i in range(0,len(mylist)):
    if mylist[i] in mydict:
      return mylist[i]
    else:
      mydict[mylist[i]]=i
  return 0

mylist = [2,1,19,7,3,19,5]
%timeit a = func(mylist)
print('The output of fun', func(mylist))


%timeit b = hashtable(mylist)
print('The output of hashtable', hashtable(mylist))
```
