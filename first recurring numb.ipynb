```
# Google question - recurring number

def rec_numb(arr):
    if len(arr) < 2:
        return arr[0]
    for i in range(len(arr)):
        for j in range(len(arr)):
            if arr[i] == arr[j] and i != j:
                return arr[i]
    return
```

```
def rec_numb2(arr):
    if len(arr) < 2:
        return arr[0]
    arr_dict = {}
    for i in range(len(arr)):
        if arr[i] not in arr_dict:
            arr_dict[arr[i]] = i
            print(arr_dict)
        else:
            return arr[i]
    return
```

```
arr1 = [2,1,1,2,3,5,1,2,4]
arr2 = [5,1,2,4]
print('The recurring number is ', rec_numb2(arr1))
```
