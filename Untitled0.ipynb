```
from random import randrange

class Pet():
    boredom_decrement = 4
    hunger_decrement = 6
    boredom_threshold = 5
    hunger_threshold = 10
    sounds = ['Mrrp']
    def __init__(self, name = "Kitty", pet_type="dog"):
        self.name = name
        self.hunger = randrange(self.hunger_threshold)
        self.boredom = randrange(self.boredom_threshold)
        self.sounds = self.sounds[:]  # copy the class attribute, so that when we make changes to it, we won't affect the other Pets in the class
        self.pet_type = pet_type

    def clock_tick(self):
        self.boredom += 1
        self.hunger += 1

    def mood(self):
        if self.hunger <= self.hunger_threshold and self.boredom <= self.boredom_threshold:
            if self.pet_type == "dog": # if the pet is a dog, it will express its mood in different ways from a cat or any other type of animal
                return "happy"
            elif self.pet_type == "cat":
                return "happy, probably"
            else:
                return "HAPPY"
        elif self.hunger > self.hunger_threshold:
            if self.pet_type == "dog": # same for hunger -- dogs and cats will express their hunger a little bit differently in this version of the class definition
                return "hungry, arf"
            elif self.pet_type == "cat":
                return "hungry, meeeeow"
            else:
                return "hungry"
        else:
            return "bored"

    def __str__(self):
        state = "     I'm " + self.name + ". "
        state += " I feel " + self.mood() + ". "
        return state

    def hi(self):
        print(self.sounds[randrange(len(self.sounds))])
        self.reduce_boredom()

    def teach(self, word):
        self.sounds.append(word)
        self.reduce_boredom()

    def feed(self):
        self.reduce_hunger()

    def reduce_hunger(self):
        self.hunger = max(0, self.hunger - self.hunger_decrement)

    def reduce_boredom(self):
        self.boredom = max(0, self.boredom - self.boredom_decrement)
```

```
def fact(n):
    #boundary condition
    if n <= 1:
        return 1
    return n * fact(n-1)
```

```
factorials = [(n, fact(n)) for n in range(9)]
print(factorials)
```

```
placement = "Beaches are cool places to visit in spring however the Mackinaw Bridge is near. Most people visit Mackinaw later since the island is a cool place to explore."

d = {}

for ch in placement:
    if ch not in d:
        d[ch] = 1
    else:
        d[ch] = d[ch] + 1
print(d)
ks = d.keys()
min_value = list(ks)[0]
print(min_value)
for k in d.keys():
    #print('k is ', k)
    if d[k] < d[min_value]:
        min_value = k
    
        
print(min_value)
```

```
product = "iphone and android phones"

lett_d = {}

for ch in product:
    if ch not in lett_d:
        lett_d[ch] = 1
    else:
        lett_d[ch] = lett_d[ch] + 1
print(lett_d)

ks = lett_d.keys()
print('ks', ks)
max_value = list(ks)[0]
#print(max_value)
for k in lett_d.keys():
    #print(k)
    if lett_d[k] > lett_d[max_value]:
        max_value = k
print(max_value)
```

```
class Point(object):
    def __init__(self, num1, num2):
        self.x = num1
        self.y = num2

    def getX(self):
        return self.x

    def getY(self):
        return self.y
    
    def DistanceFromOrigin(self):
        #print('x and y', self.x, self.y)
        return (self.x **2 + self.y **2 ) ** 0.5
    
    def __str__(self):
        return '{} and {} are x and y'.format(self.x, self.y)

    
def find_distance(p1,p2):
    x_diff = p1.getX() - p2.getX()
    y_diff = p1.getY() - p2.getY()
    #print('x and y are ', x_diff, y_diff)
    return (x_diff **2 + y_diff **2) ** 0.5


p1 = Point(6, 10)
p2 = Point( 4,3)
print(p1.x)
#print(find_distance(p1,p2))
#print(p1)
assert p2.DistanceFromOrigin() == 5



```

```

class AppleBasket():
    ''' Apple Basket '''
    def __init__(self, c, q):
        self.apple_color = c
        self.apple_quantity = q
        
    def __str__(self):
        return 'A basket of {1} {0} apples.'.format(self.apple_color, self.apple_quantity)
    
    def increase(self):
        self.apple_quantity += 1
        
a1 = AppleBasket('red', 4)

print(a1)
a1.increase()
print(a1)
a1.c = 'purple'
print(a1)

        
        
```
