```python
# let's create a class of queue

class MyQueue(object):
    def __init__(self):
        self.q = []

    def IsEmpty(self):
        return len(self.q) == 0

    def Size(self):
        return len(self.q)

    def Enqueue(self, n):
        self.n = n
        self.q.insert(0,n)

    def Dequeue(self):
        return self.q.pop()

```

```python
q = MyQueue()
print(q.IsEmpty())
print(q.Size())
q.Enqueue(5)
q.Enqueue(9)
q.Enqueue(24)
print(q.IsEmpty())
print(q.Size())
print(q.Dequeue())
print(q.Dequeue())
```

```python
# let's create a class of queue

class MyStack(object):
    def __init__(self):
        self.q = []

    def IsEmpty(self):
        return len(self.q) == 0

    def Size(self):
        return len(self.q)

    def Enqueue(self, n):
        self.n = n
        self.q.append(n)

    def Dequeue(self):
        return self.q.pop()
```

```python
q = MyStack()
print(q.IsEmpty())
print(q.Size())
q.Enqueue(5)
q.Enqueue(9)
q.Enqueue(24)
print(q.IsEmpty())
print(q.Size())
print(q.Dequeue())
print(q.Dequeue())
```

```python
class Deque(object):
    def __init__(self):
        self.q = []
    
    def IsEmpty(self):
        return len(self.q) == 0

    def Size(self):
        return len(self.q)

    def AddFront(self, n):
        self.q.append(n)

    def AddRear(self, n):
        self.q.insert(0, n)

    def RemoveFront(self):
        return self.q.pop()

    def RemoveRear(self):
        return self.q.pop(0)

```

```python
d = Deque()
print(d.IsEmpty())
print(d.Size())
d.AddFront(5)
d.AddFront(9)
d.AddFront(24)
print(d.IsEmpty())
print(d.Size())
print(d.RemoveFront())
print(d.RemoveRear())
```
