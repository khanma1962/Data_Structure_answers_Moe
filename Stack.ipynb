```
#stack

class Stack():

    def __init__(self):
        self.length = 0
        self.stack = []

    def __str__(self):
        return str(self.__dict__)

    def is_empty(self):
        return self.stack == []
        
    def peek(self):
        return self.stack[self.length - 1]
    
    def append(self, num):
        self.stack.append(num)
        self.length += 1
        return

    def pop(self):
        poped_item = self.stack.pop()
        self.length -= 1
        return poped_item
    
    def is_length(self):
        return len(self.stack)
```

```
s = Stack()
s.is_empty()
s.append(5)
s.append(7)
print('Peek at 7', s.peek())
s.append(25)

print('length is ', s.length)
print('stack is ', s.stack)
retrieve_item = s.pop()
print('retrieve item', retrieve_item)
print('length is ', s.length)
print('stack is ', s.stack)
s.is_length()
```
