```python
def unique_char(s):
    if len(s) == 0:
        return print('String is empty')
    d1 = {}
    for char in s:
        if char not in d1:
            d1[char] = 1
        else:
            d1[char] += 1
        # print('d1 is', d1)
    
    for k,v in d1.items():
        if v >= 2:
            return False
    return True
    
```

```python
unique_char('')
print(unique_char('abcddde'))
print(unique_char('abcdde'))
print(unique_char('abcde'))
```
