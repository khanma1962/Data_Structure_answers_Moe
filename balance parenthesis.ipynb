```python
def balance_parenthesis(s):
    if len(s) == 0: # boundary condition
        return print('it is an empty string')

    if (len(s) % 2) != 0: # boundary condition
        return False
    
    d1 = {}
    s = s.replace(' ', '')
    for char in s:
        if char not in d1:
            d1[char] = 1
        else:
            d1[char] += 1
        print('d1 is', d1)



```

```python
balance_parenthesis('[}{]')
```
