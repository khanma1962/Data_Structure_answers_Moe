```
class Graph:

    def __init__(self):
        self.number_of_nodes = 0
        self.adjacent_list = {}

    def __str__(self):
        return str(_dict__)

    def add_vertex(self, node):
        self.adjacent_list[node] = []
        self.number_of_nodes += 1
        # print('adj', self.adjacent_list)
        # print('no of nodes', self.number_of_nodes)

    def add_edges(self, node1, node2):
        self.adjacent_list[node1].append(node2)
        self.adjacent_list[node2].append(node1)
        # print('adj', node1, node2,self.adjacent_list)
        # print('no of nodes', self.number_of_nodes)
    def show_connection(self):
        print('adj', self.adjacent_list)
        for x in self.adjacent_list:
            print('\n',x, end= ' --- >')
            for i in self.adjacent_list[x]:
                print(i, end='  ')
```

```
g = Graph()
g.add_vertex('0')
g.add_vertex('1')
g.add_vertex('2')
g.add_vertex('3')
g.add_vertex('4')
g.add_vertex('5')
g.add_vertex('6')
g.add_edges('3', '1')
g.add_edges('3', '4')
g.add_edges('4', '2') 
g.add_edges('4', '5') 
g.add_edges('1', '2') 
g.add_edges('1', '0') 
g.add_edges('0', '2') 
g.add_edges('6', '5')
g.show_connection()
```
