```python
# 2 sum
def pair_sum1(arr, sum): # o(n2)
    if len(arr) == 1: # boundary condition
        return len(arr)
    for i in range(len(arr)):
        for j in range(len(arr)):
            if i != j and arr[i] + arr[j] == sum:
                print(arr[i], arr[j])

def pair_sum2(arr, sum): # o(n)
    d1 = {}
    for n in arr:
        if (sum - n) in d1:
            print('pair found', n, d1[sum-n])
        else:
            d1[n] = 1

```

```python
pair_sum2([1,3,2,2], 4)
```
