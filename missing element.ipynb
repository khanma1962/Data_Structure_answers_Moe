```python
def missing_elem(arr1, arr2):
    if len(arr1) != len(arr2):
        print('The length is not equal')
    d1 = {} 
    for elem in arr2:
        if elem in d1:
            d1[elem] += 1
        else:
            d1[elem] = 1

        print('d1 is', d1)
    
    for elem in arr1:
        if elem not in d1:
            print('missing element is ', elem)

```

```python
a = [1,2, 3,4,5, 6, 7]
b = [1,2, 7,3, 6,4]
missing_elem(a,b)
```
