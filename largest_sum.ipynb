```python
def largest_sum(arr):
    if len(arr) == 1: # boundary condition
        return arr
    # make max_sum, curr_sum as equal to arr[0]
    max_sum = curr_sum = arr[0]
    for num in arr[1:]:
        # check if num or num+ curr_sum is greated
        print('checking num', num)
        curr_sum = max(num, num + curr_sum)
        print('curr_sum is', curr_sum)

        # assign max_sum to the largest of curr/max
        max_sum = max(curr_sum, max_sum)
        print('max_sum is', max_sum)

    return max_sum
```

```python
largest_sum([1,2,-1,3,4,10,10,-10,-1])
```
