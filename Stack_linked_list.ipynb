```
# stack with linked list

class Node():
    def __init__(self,data):
        self.data = data
        self.next = None

class Stack_linked_list():
    def __init__(self):
        self.top = None
        self.bottom = None
        self.length = 0

    def is_empty(self):
        return self.length == 0

    def peek(self):
        return self.top.data
    
    def push(self,num):
        new_node = Node(num)
        if self.bottom == None:
            print('list is empty adding one node')
            self.top = new_node
            self.bottom = new_node
            
        else:
            print('list is NOT empty adding one node')
            new_node.next = self.top
            self.top = new_node
            # self.top.next = temp_storage
            print('top:', self.top.data, 'top next :', self.top.next.data)
        
        self.length += 1
        return
    
    def pop(self):
        if self.bottom == None:
            return None
        print('popping cell now')
        holding_cell = self.top
        self.top = self.top.next
        return holding_cell

```

```
s = Stack_linked_list()
print(s.is_empty())
# print(s.peek())
s.push('google')
print(s.is_empty())
print(s.peek())
s.push('udemy')
print(s.is_empty())
print(s.peek())
s.push('microsoft')
print(s.is_empty())
print(s.peek())
popped_cell = s.pop()
print('popped cell', popped_cell.data)
```
